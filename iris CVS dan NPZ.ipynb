```python
import numpy as np

# Fungsi aktivasi step
def step_function(x):
    return 1 if x >= 0 else 0

# Kelas Perceptron
class Perceptron:
    def __init__(self, input_size, learning_rate=0.1, epochs=10):
        self.weights = np.zeros(input_size + 1)  # +1 untuk bias
        self.learning_rate = learning_rate
        self.epochs = epochs

    def predict(self, x):
        x = np.insert(x, 0, 1)  # Tambahkan bias input (1) di awal
        z = np.dot(self.weights, x)
        return step_function(z)

    def train(self, X, y):
        for epoch in range(self.epochs):
            print(f"Epoch {epoch+1}")
            for i in range(len(X)):
                x = np.insert(X[i], 0, 1)  # Tambahkan bias input
                y_pred = self.predict(X[i])
                error = y[i] - y_pred
                self.weights += self.learning_rate * error * x
                print(f"  Input: {X[i]}, Target: {y[i]}, Prediksi: {y_pred}, Bobot: {self.weights}")

# Data input (fungsi logika AND)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 0, 0, 1])  # Output target untuk logika AND

# Inisialisasi dan latih model
p = Perceptron(input_size=2)
p.train(X, y)

# Uji hasil akhir
print("\n=== Pengujian ===")
for x in X:
    print(f"Input: {x}, Prediksi: {p.predict(x)}")
```
